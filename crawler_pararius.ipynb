## CRAWLER

```python
import pandas as pd
import cfscrape
from lxml import etree
from datetime import datetime
pd.set_option('display.max_colwidth', -1)
```

```python
header = {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
          'Accept-Encoding': 'gzip, deflate, sdch',
          'Accept-Language' : 'nl-NL,nl;q=0.8,en-US;q=0.6,en;q=0.4',
          'Cache-Control' : 'max-age=0',
          'Connection': 'keep-alive',
          'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.81 Safari/537.36'}
```

```python
url = "https://www.pararius.nl/huurwoningen/nieuwegein/0-1400/100m2/woonhuis"
```

```python
scraper = cfscrape.create_scraper()  
scraped_html=scraper.get(url,headers=header).content
print(scraped_html.decode("utf-8") )
```

```python
html = etree.HTML(scraped_html)
```

```python
df_all_links = df_all_links = pd.DataFrame(html.xpath("//div[@class='details']/h2/a/@href"))
df_all_links[0] = "https://www.pararius.nl"+df_all_links[0]
```

```python
i=0
address = ""
for entry in html.xpath("//div[@class='details']/ul[@class='breadcrumbs']/li/text()"):

    if i % 2 != 0:
        address = address + ' ' +entry.strip()
        print(address)
        address=""
    if i % 3 == 0:
        next
    i = i+1
```
